````python
# Import Dependencies
import sqlalchemy
import pandas as pd
from sqlalchemy import create_engine
from config import password
````

## EXTRACTION (Finding Data)

````python
# Store New York CSV into a dataframe
csv_file = "./Resources/newyork_housing.csv"
ny_area_df = pd.read_csv(csv_file)
ny_area_df
````

````python
# Store Austin CSV into a dataframe
csv_file = "./Resources/austin_housing.csv"
autin_df = pd.read_csv(csv_file)
autin_df.head()
````

### TRANSFORMATIONS (Data Cleanup & Analysis)

````python
# Create new data with select columns
ny_clean_df = ny_area_df[["bedrooms", "price", "address/streetAddress", "address/city"]].copy()
ny_clean_df = ny_clean_df.rename(columns={'bedrooms': 'num_bedrooms', 'address/streetAddress': 'street_address', 'address/city': 'city'})
ny_clean_df
````

````python
# Drop NaN values
ny_clean_df.dropna()
````

````python
# Find average price grouped by number of bedrooms
avg_ny_price = ny_clean_df.groupby('num_bedrooms').mean()
avg_ny_price
````

````python
# Convert 'price' from string to float type
avg_ny = avg_ny_price["price"].astype(float)
avg_ny
````

````python
# Find median price grouped by number of bedrooms
median_ny_price = ny_clean_df.groupby('num_bedrooms').median()
median_ny_price
````

````python
# Convert 'price' from string to float type
median_ny = median_ny_price["price"].astype(float)
median_ny
````

````python
# Merge the average and median price tables
ny_stats = pd.merge(avg_ny_price, median_ny_price, how="left", on="num_bedrooms")
ny_stats = ny_stats.rename(columns={'price_x': 'Average Price', 'price_y': 'Median Price'})
ny_stats.head()
````

````python
# Dataframe formatting
ny_stats["Average Price"] = ny_stats["Average Price"].map("${:,.2f}".format)
ny_stats.head()
````

````python
# Dataframe formatting
ny_stats["Median Price"] = ny_stats["Median Price"].map("${:,.2f}".format)
ny_stats.head()
````

````python
# Create new data with select columns for Austin dataframe
austin_clean_df = autin_df[["bedrooms", "latest_price", "address_streetAddress", "address_city"]].copy()
austin_clean_df = austin_clean_df.rename(columns={'bedrooms': 'num_bedrooms', 'latest_price': 'price', 'address_streetAddress': 'street_address', 'address_city': 'city'})
austin_clean_df
````

````python
# Drop NaN values
austin_clean_df.dropna()
````

````python
# Find average price grouped by number of bedrooms
avg_austin_price = austin_clean_df.groupby('num_bedrooms').mean()
avg_austin_price
````

````python
# Convert 'latest_price' from string to float type
austin_avg = avg_austin_price["price"].astype(float)
austin_avg
````

````python
# Find median price grouped by number of bedrooms
median_austin_price = austin_clean_df.groupby('num_bedrooms').median()
median_austin_price
````

````python
# Convert 'price' from string to float type
austin_median = median_austin_price["price"].astype(float)
austin_median
````

````python
# Merge the Austin average and median price tables
austin_stats = pd.merge(avg_austin_price, median_austin_price, how="left", on="num_bedrooms")
austin_stats = austin_stats.rename(columns={'price_x': 'Average Price', 'price_y': 'Median Price'})
austin_stats.head()
````

````python
# Dataframe formatting
austin_stats["Average Price"] = austin_stats["Average Price"].map("${:,.2f}".format)
austin_stats.head()
````

````python
# Dataframe formatting
austin_stats["Median Price"] = austin_stats["Median Price"].map("${:,.2f}".format)
austin_stats.head()
````

## Analysis

````python
# Analyzing New York vs. Austin housing prices
ny_austin_df = pd.merge(ny_stats, austin_stats, how='left', on='num_bedrooms')
ny_austin_df = ny_austin_df.rename(columns={'Average Price_x': 'New York avg price', 'Median Price_x': 'New York median price', 'Average Price_y': 'Austin avg price', 'Median Price_y': 'Austin median price'})

ny_austin_df.head()
````

#### Create a schema for where data will be loaded
```sql
CREATE TABLE newyork_housing (
	num_bedrooms INT,
	price NUMERIC,
	street_address TEXT,
	city TEXT
);

CREATE TABLE austin_housing (
	num_bedrooms INT,
	price NUMERIC,
	street_address TEXT,
	city TEXT
);

CREATE TABLE yelp_users (
    user_id INT PRIMARY KEY,
    user_name TEXT,
    yelping_since DATE,
    review_count INT,
    average_stars INT
);

SELECT * FROM newyork_housing
SELECT * FROM austin_housing
SELECT * FROM yelp_users

```

````python
# Create engine and connection to ETL_city_housing db.
engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/ETL_city_housing')
connection = engine.connect()
````

````python
# Check the tables
engine.table_names()
````

### LOADING

````python
# Use pandas to load csv converted to DF into database
ny_clean_df.to_sql(name="newyork_housing", con=engine, if_exists='append', index=False)
````

````python
# Use pandas to load csv converted to DF into database
austin_clean_df.to_sql(name="austin_housing", con=engine, if_exists='append', index=False)
````

````python
# Use pandas to load json converted to DF into database 
yelp_users_df.to_sql(name='yelp_users', con=engine, if_exists="append", index=False)
````

````python
# Calculate average New York housing price grouped by number of bedrooms
avg_ny_price = pd.read_sql('SELECT "num_bedrooms", AVG("price") AS "Average Housing Price" FROM newyork_housing GROUP BY "num_bedrooms" ORDER BY "num_bedrooms"', con=engine)
avg_ny_price
````

````python
# Calculate average Austin housing price grouped by number of bedrooms
avg_austin_price = pd.read_sql('SELECT "num_bedrooms", AVG("price") AS "Austin Average Price" FROM austin_housing GROUP BY "num_bedrooms" ORDER BY "num_bedrooms"', con=engine)
avg_austin_price
````

````python
# Find top Yelp users by 'review counts'
top_yelp_users = pd.read_sql('SELECT user_name, SUM(review_count) AS "Review Count" FROM yelp_users GROUP BY user_name, review_count ORDER BY review_count DESC', con=engine)
top_yelp_users
````
