````python
# Import Dependencies
import sqlalchemy
import pandas as pd
from sqlalchemy import create_engine
from config import password
````

````python
# Store JSON data into DataFrame 
json_file = ('./Resources/yelp_academic_dataset_user.json')
yelp_df = pd.read_json(json_file, lines=True)
````

````python
yelp_df
````

````python
# Create new data with select columns
yelp_users_df = yelp_df[["user_id", "name", 'yelping_since', 'review_count', 'average_stars']].copy()
yelp_users_df = yelp_users_df.rename(columns={'name': 'user_name'})
yelp_users_df
````

````python
user_review_df = yelp_users_df.dropna()
user_review_df
````

#### Create a schema for where data will be loaded

```sql
CREATE TABLE yelp_users (
    user_id INT PRIMARY KEY,
    user_name TEXT,
    yelping_since DATE,
    review_count INT,
    average_stars INT
);

SELECT * FROM yelp_users

```

````python
# Create engine and connection to ETL_city_housing db.
engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/ETL_yelp_users')
connection = engine.connect()
````

````python
# Check the tables
engine.table_names()
````

````python
# Use pandas to load json converted to DF into database 
user_review_df.to_sql(name='yelp_users', con=engine, if_exists="append", index=False)
````

````python
# Confirm data is in the newyork_housing table
pd.read_sql_query('SELECT * FROM yelp_users', con=engine)
````

````python
# Find top 20 Yelp Users by 'review counts'
top_yelp_users = pd.read_sql('SELECT user_id, user_name, SUM(review_count) AS "Review Count" FROM yelp_users GROUP BY user_id, user_name, review_count ORDER BY review_count DESC', con=engine)
top_yelp_users.head(20)
````

````python
# Find first 50 Yelp Users based on 'yelping_since' values
first50_yelp_users = pd.read_sql('SELECT user_id, user_name, yelping_since FROM yelp_users GROUP BY user_id, user_name, yelping_since ORDER BY yelping_since', con=engine)
first50_yelp_users.head(50)
````
